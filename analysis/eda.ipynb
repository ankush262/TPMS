```python
#import libraries some of them are listed in  requirements.txt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
df = pd.read_csv(r'E:\TPMS\data\raw\configuration_data.csv')
df.head()
```

```python
df.info()
```

```python
print(df.isnull().sum())
print(df.duplicated().sum())
df.shape
```

```python
Object_columns = df[['max_load_single_kg','section_width_mm','overall_diameter_mm','tyre_life']]
Object_columns.head()
```

```python
#striping units form the object columns and converting them to numeric values
Object_columns = Object_columns.apply(lambda x: x.str.replace(r'[^\d\.]', '', regex=True))
Object_columns = Object_columns.apply(pd.to_numeric)
print(Object_columns.dtypes)
print(Object_columns.head())
```

```python
#split the axle_configuration column into two separate columns for front and rear axle configurations
df[['total_wheels','driven_wheels']] = df['axle_configuration'].str.split('x', expand=True)
#convert the extraceted strings into int 64
df['total_wheels'] = pd.to_numeric(df['total_wheels'], errors='coerce')
df['driven_wheels'] = pd.to_numeric(df['driven_wheels'], errors='coerce')
df.head()
```

```python
#updating the original datafram with the cleaned numeric columns
df[['max_load_single_kg','section_width_mm','overall_diameter_mm','tyre_life']] = Object_columns
#now separating numeric columns and categorical columns
numeric_columns = df.select_dtypes(include=[np.number])
categorical_columns = df.select_dtypes(exclude=[np.number])
print("Numeric Columns:")
print(numeric_columns.columns)
print("\nCategorical Columns:")
print(categorical_columns.columns)
```

```python
# check current column names before calculation
print(df.columns)

# compute derived features using actual column names
# driven ratio is percentage of wheels that are driven
df['driven_ratio'] = df['driven_wheels'] / df['total_wheels']

# ensure gvwr and curb weight are numeric
for col in ['gvwr_kg', 'curb_weight']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# load per wheel (subtract curb weight from GVWR)
df['load_per_wheel'] = (df['gvwr_kg'] - df['curb_weight']) / df['total_wheels']

# drive stress factor
# avoid division by zero by filling driven_ratio zeros with nan
df['drive_stress_factor'] = df['load_per_wheel'] / df['driven_ratio'].replace({0: np.nan})
```

```python
#updating this new datafram and saving it into E:\TPMS\data\processed location
df.to_csv(r'E:\TPMS\data\processed\cleaned_configuration_data.csv', index=False)
```
